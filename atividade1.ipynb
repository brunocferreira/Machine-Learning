```python
#--- Libraries ---#
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('./lib/iris/iris.data') # https://archive.ics.uci.edu/dataset/53/iris
iris.head()
```

```python
iris.columns = [
  'sepal_length_in_cm',
  'sepal_width_in_cm',
  'petal_length _in_cm',
  'petal_width_in_cm',
  'class'
  ]
iris
```

```python
classes, tipos = iris.iloc[:, -1].factorize()
iris['classe'] = classes
iris
```

```python
sns.pairplot(iris, hue = 'class', height=1.5)
```

```python
x_iris = iris.drop(['class', 'classe'], axis=1)
x_iris.shape
```

```python
y_iris = iris['classe']
y_iris.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.naive_bayes import GaussianNB as NB
```

```python
classificador = 3
```

```python
#from sklearn.neighbors import KNeighborsClassifier as KNN

knn = KNN(n_neighbors=classificador)

knn.fit(x_iris,y_iris)
```

```python
find_array = np.random.rand(1,4)
find_array = find_array*100
found = knn.predict(find_array)
print(f'find_array: {find_array}')
print(f'found: {found}')
```

```python
# from sklearn.naive_bayes import GaussianNB as NB
clf = NB()
clf.fit(x_iris,y_iris)
```

```python
clf.predict(find_array)
```
